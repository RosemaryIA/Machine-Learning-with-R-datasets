```python
import pandas as pd

# Defini√ß√£o do dicion√°rio de perguntas (fora da fun√ß√£o)
perguntas = {
    "experiencia": "Em uma escala de 0 a 10, sendo 0 ‚Äúnada satisfeito‚Äù e 10 ‚Äúextremamente satisfeito‚Äù, como voc√™ avalia sua experi√™ncia geral com a WeLovePizza hoje?",
    "pontos_fortes": "O que voc√™ mais gosta na WeLovePizza? (Pode escolher mais de uma op√ß√£o)\n a) Sabor das pizzas\n b) Variedade do card√°pio\n c) Qualidade dos ingredientes\n d) Pre√ßo\n e) Atendimento\n f) Tempo de entrega\n g) Ambiente da loja\n h) Outro: __________",
    "pontos_a_desenvolver": "O que poder√≠amos melhorar na WeLovePizza? (Pode escolher mais de uma op√ß√£o)\n a) Sabor das pizzas\n b) Variedade do card√°pio\n c) Qualidade dos ingredientes\n d) Pre√ßo\n e) Atendimento\n f) Tempo de entrega\n g) Ambiente da loja\n h) Outro: __________",
    "frequencia_compra": "Nos √∫ltimos 3 meses, com que frequ√™ncia voc√™ comprou pizzas?\n a) Somente na WeLovePizza\n b) Mais na WeLovePizza do que na concorr√™ncia\n c) Mais na concorr√™ncia do que na WeLovePizza\n d) Somente na concorr√™ncia",
    "sentimento": "Qual palavra melhor descreve o que voc√™ sente pela WeLovePizza?\n a) Satisfa√ß√£o\n b) Confian√ßa\n c) Indiferen√ßa\n d) Frustra√ß√£o\n e) Outro: __________",
    "sugestoes": "Tem alguma sugest√£o para melhorarmos ainda mais?"
}
```

```python
 # Cria um DataFrame vazio com as colunas das perguntas
df = pd.DataFrame(columns=list(perguntas.keys()))
```

```python
def chatbot_welovepizza(df):
    # Cria um DataFrame vazio com as colunas das perguntas

    # Mensagens iniciais
    print("Ol√°! Agradecemos sua visita √† WeLovePizza. Queremos saber sua opini√£o para melhorarmos ainda mais!")
    print("Respondendo a esta pesquisa r√°pida, voc√™ concorre a 3 vinhos de sua escolha dentre uma sele√ß√£o especial. üç∑")
    print("Para conferir as op√ß√µes, acesse este link: [Link para a carta de vinhos]\n")

    # Lista para armazenar as respostas da itera√ß√£o atual
    respostas_iteracao = []

    # Loop para exibir as perguntas sequencialmente
    for key, pergunta in perguntas.items():
        if key == "experiencia":
            # Valida√ß√£o espec√≠fica para a quest√£o 1
            while True:
                try:
                    resposta = int(input(pergunta + "\nResposta (0-10): "))
                    if 0 <= resposta <= 10:
                        break
                    else:
                        print("Por favor, digite um n√∫mero entre 0 e 10.")
                except ValueError:
                    print("Entrada inv√°lida. Digite um n√∫mero.")
        else:
            # Valida√ß√£o para as demais quest√µes (com m√∫ltiplas escolhas)
            while True:
                resposta = input(pergunta + "\nResposta (separe as op√ß√µes por ponto e v√≠rgula para mais de uma resposta para a mesma pergunta): ").lower()
                opcoes_escolhidas = resposta.split(";")  # Separar as op√ß√µes
                opcoes_validas = ["a", "b", "c", "d", "e", "f", "g", "h"]
                if all(opcao in opcoes_validas for opcao in opcoes_escolhidas):  # Verificar se todas as op√ß√µes s√£o v√°lidas
                    break
                else:
                    print("Op√ß√£o inv√°lida. Por favor, escolha letras de 'a' a 'h', separadas por ponto e v√≠rgula.")

        # Adiciona a resposta √† lista da itera√ß√£o atual
        respostas_iteracao.append(resposta)

        print()  # Linha em branco para separar as perguntas

    # Adiciona as respostas da itera√ß√£o atual ao DataFrame como uma nova linha
    df = pd.concat([df, pd.DataFrame([respostas_iteracao], columns=df.columns)], ignore_index=True)

    # Mensagem final
    print("Muito obrigado pela sua participa√ß√£o! Voc√™ j√° est√° concorrendo aos 3 vinhos. O resultado ser√° divulgado em [data] em nossas redes sociais.")

    # Imprime o DataFrame
    print(df)

    return df

#if __name__ == "__main__":  chatbot_welovepizza(df)
```

###Simulando respostas para 135 question√°rios para realizar a an√°lise de dados e construir o dataframe

####Intru√ß√µes para a contru√ß√£o do dataframe: prompt no AIStudio:
 Importante que as respostas apresentem :


1.   "Em uma escala de 0 a 10, sendo 0 ‚Äúnada satisfeito‚Äù e 10 ‚Äúextremamente satisfeito‚Äù, como voc√™ avalia sua experi√™ncia geral com a WeLovePizza hoje?",  - m√©dia dos question√°rios 4,5:

2.   que as respostas denotem insatisfa√ß√£o dos clientes quanto a variedade do card√°pio,  atendimento e tempo de entrega

3.   que o campo outros esteja preenchido para ao menos 20% dos question√°rio e que desses 60 das  respostas denotem sentimento de insatisfa√ß√£o com os produtos e servi√ßo e fuga dos clientes para a concorr√™ncia.

4.   a quest√£o : "Tem alguma sugest√£o para melhorarmos ainda mais?", tenha um preenchimento de 30% com sugest√µes de melhorias em geral dos produtos, servi√ßos

5.   O C√≥digo deve gerar um dataframe onde cada linha √© um question√°rio e as colunas s√£o as quest√µes


```python
import pandas as pd
import random

def gerar_resposta(tipo, insatisfeito=False):
    if tipo == "experiencia":
        # Distribui√ß√£o para m√©dia 4.5, com vi√©s para baixo se insatisfeito
        if insatisfeito:
            return random.choices([0, 1, 2, 3, 4], weights=[3, 3, 2, 1, 1])[0]
        else:
            return random.choices([3, 4, 5, 6, 7], weights=[1, 2, 3, 2, 1])[0]
    elif tipo in ["pontos_fortes", "pontos_a_desenvolver"]:
        opcoes = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h']
        n = random.randint(1, 3)
        if tipo == "pontos_a_desenvolver" or insatisfeito:
            # Garante que 'b', 'e', 'f' estejam presentes se insatisfeito
            escolhas = random.sample(opcoes, n)
            if 'b' not in escolhas: escolhas.append('b')
            if 'e' not in escolhas: escolhas.append('e')
            if 'f' not in escolhas: escolhas.append('f')
            return ';'.join(escolhas)
        else:
            return ';'.join(random.sample(opcoes, n))
    elif tipo == "frequencia_compra":
        if insatisfeito:
            return random.choices(['c', 'd'], weights=[3, 1])[0]  # Vi√©s para concorr√™ncia
        else:
            return random.choices(['a', 'b', 'c', 'd'], weights=[2, 4, 2, 1])[0]
    elif tipo == "sentimento":
        if insatisfeito:
            return random.choices(['c', 'd', 'e'], weights=[1, 3, 1])[0]  # Vi√©s para insatisfa√ß√£o
        else:
            return random.choices(['a', 'b', 'c'], weights=[4, 3, 1])[0]
    else:  # tipo == "sugestoes"
        return "Sugest√£o de melhoria {}".format(random.randint(1, 10))

questionarios = []
for i in range(135):
    insatisfeito = i >= 108  # 20% dos question√°rios (27) ser√£o de clientes insatisfeitos
    questionario = {
        "experiencia": gerar_resposta("experiencia", insatisfeito),
        "pontos_fortes": gerar_resposta("pontos_fortes", insatisfeito),
        "pontos_a_desenvolver": gerar_resposta("pontos_a_desenvolver", insatisfeito),
        "frequencia_compra": gerar_resposta("frequencia_compra", insatisfeito),
        "sentimento": gerar_resposta("sentimento", insatisfeito),
        "sugestoes": gerar_resposta("sugestoes") if random.random() < 0.3 else ""  # 30% preenchido
    }
    if insatisfeito:
        questionario["pontos_a_desenvolver"] += ";Outro: Insatisfeito com X"
    elif random.random() < 0.2:  # 20% de "Outros" preenchido
        questionario["pontos_a_desenvolver"] += ";Outro: Algo espec√≠fico"
    questionarios.append(questionario)

df = pd.DataFrame(questionarios)

# Exibir o DataFrame (opcional)
#print(df)

# C√°lculo da m√©dia da experi√™ncia
media_experiencia = df["experiencia"].mean()
print("\nM√©dia da experi√™ncia:", media_experiencia)
```

```python
import pandas as pd
import random

# Dicion√°rios de mapeamento
pontos_fortes_dict = {
    'a': 'Sabor das pizzas',
    'b': 'Variedade do card√°pio',
    'c': 'Qualidade dos ingredientes',
    'd': 'Pre√ßo',
    'e': 'Atendimento',
    'f': 'Tempo de entrega',
    'g': 'Ambiente da loja'
}

pontos_a_desenvolver_dict = {
    'a': 'Sabor das pizzas',
    'b': 'Variedade do card√°pio',
    'c': 'Qualidade dos ingredientes',
    'd': 'Pre√ßo',
    'e': 'Atendimento',
    'f': 'Tempo de entrega',
    'g': 'Ambiente da loja'
}

frequencia_compra_dict = {
    'a': 'Somente na WeLovePizza',
    'b': 'Mais na WeLovePizza do que na concorr√™ncia',
    'c': 'Mais na concorr√™ncia do que na WeLovePizza',
    'd': 'Somente na concorr√™ncia'
}

sentimento_dict = {
    'a': 'Satisfa√ß√£o',
    'b': 'Confian√ßa',
    'c': 'Indiferen√ßa',
    'd': 'Frustra√ß√£o'
}

def traduzir_respostas(respostas_codificadas, dicionario):
    respostas = []
    for resposta in respostas_codificadas.split(';'):
        if resposta in dicionario:
            respostas.append(dicionario[resposta])
    return respostas

questionarios = []
for i in range(135):
    insatisfeito = i >= 108
    questionario = {
        "experiencia": gerar_resposta("experiencia", insatisfeito),
        "pontos_fortes": gerar_resposta("pontos_fortes", insatisfeito),
        "pontos_a_desenvolver": gerar_resposta("pontos_a_desenvolver", insatisfeito),
        "frequencia_compra": gerar_resposta("frequencia_compra", insatisfeito),
        "sentimento": gerar_resposta("sentimento", insatisfeito),
        "sugestoes": gerar_resposta("sugestoes") if random.random() < 0.3 else ""
    }
    # ... (mesma l√≥gica para preencher "Outro" e "sugestoes")
    questionarios.append(questionario)

df = pd.DataFrame(questionarios)

# Aplicar os dicion√°rios para traduzir as respostas
df['pontos_fortes_detalhados'] = df['pontos_fortes'].apply(lambda x: traduzir_respostas(x, pontos_fortes_dict))
df['pontos_a_desenvolver_detalhados'] = df['pontos_a_desenvolver'].apply(lambda x: traduzir_respostas(x, pontos_a_desenvolver_dict))
df['frequencia_compra_detalhada'] = df['frequencia_compra'].apply(lambda x: traduzir_respostas(x, frequencia_compra_dict)[0] if traduzir_respostas(x, frequencia_compra_dict) else "")
df['sentimento_detalhado'] = df['sentimento'].apply(lambda x: traduzir_respostas(x, sentimento_dict)[0] if traduzir_respostas(x, sentimento_dict) else "")

# Exibir o DataFrame com as novas colunas
print(df)

# C√°lculo da m√©dia da experi√™ncia
media_experiencia = df["experiencia"].mean()
print("\nM√©dia da experi√™ncia:", media_experiencia)
```

##Data ***Analysis***

```python
# 1. Experi√™ncia Geral
media_experiencia = df["experiencia"].mean()

# 2. Pontos Fortes
todos_pontos_fortes = []
for lista_pontos_fortes in df['pontos_fortes_detalhados']:
    todos_pontos_fortes.extend(lista_pontos_fortes)
fortes_counts = pd.Series(todos_pontos_fortes).value_counts()

# 3. Pontos a Desenvolver
todos_pontos_a_desenvolver = []
for lista_pontos_a_desenvolver in df['pontos_a_desenvolver_detalhados']:
    todos_pontos_a_desenvolver.extend(lista_pontos_a_desenvolver)
fracos_counts = pd.Series(todos_pontos_a_desenvolver).value_counts()

# 4. Frequ√™ncia de Compra
frequencia_counts = df['frequencia_compra_detalhada'].value_counts()

# 5. Sentimento
sentimento_counts = df['sentimento_detalhado'].value_counts()

# --- Gera√ß√£o do Relat√≥rio ---

print("----- Relat√≥rio da Pesquisa de Satisfa√ß√£o WeLovePizza -----")

print("\n1. Experi√™ncia Geral:")
print(f"  M√©dia da Experi√™ncia: {media_experiencia:.2f}")

print("\n2. Pontos Fortes:")
print(fortes_counts)

print("\n3. Pontos a Desenvolver:")
print(fracos_counts)

print("\n4. Frequ√™ncia de Compra:")
print(frequencia_counts)

print("\n5. Sentimento:")
print(sentimento_counts)
```

```python
from xml.dom.minidom import getDOMImplementation

def criar_html(titulo, conteudo):
  """Cria um arquivo HTML usando a biblioteca DOM."""
  impl = getDOMImplementation()
  doc = impl.createDocument(None, "html", None)

  html = doc.documentElement
  head = doc.createElement("head")
  title = doc.createElement("title")
  title.appendChild(doc.createTextNode(titulo))
  head.appendChild(title)
  html.appendChild(head)

  body = doc.createElement("body")
  h1 = doc.createElement("h1")
  h1.appendChild(doc.createTextNode(titulo))
  body.appendChild(h1)
  p = doc.createElement("p")
  p.appendChild(doc.createTextNode(conteudo))
  body.appendChild(p)
  html.appendChild(body)

  return doc.toprettyxml()

# Exemplo de uso
titulo = "Meu Relat√≥rio"
conteudo = "Este √© o conte√∫do do meu relat√≥rio."
html_string = criar_html(titulo, conteudo)

# Salvar o HTML em um arquivo
with open("relatorio.html", "w") as f:
  f.write(html_string)
```
