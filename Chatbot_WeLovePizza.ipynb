Explicação:
1.	Função chatbot_welovepizza(): Contém toda a lógica do chatbot.
2.	Mensagens iniciais: Exibe as boas-vindas e o incentivo do sorteio.
3.	Dicionário perguntas: Armazena as perguntas da pesquisa com suas opções de resposta.
4.	Dicionário respostas_usuario: Simula as respostas do usuário (em um cenário real, você usaria input() para obter as respostas).
5.	Loop de perguntas: Itera sobre as perguntas, exibe cada uma e a respectiva resposta simulada.
6.	Mensagem final: Agradece a participação e informa sobre o sorteio.
Para tornar o chatbot interativo:
1.	Substitua as respostas simuladas em respostas_usuario por input() para que o usuário possa digitar suas respostas.
2.	Implemente lógica para validar as respostas (por exemplo, garantir que a resposta para a escala de 0 a 10 seja um número válido).
3.	Utilize uma biblioteca de interface gráfica (como Tkinter ou PyQt) para criar uma interface visual para o chatbot.
4.	Integre o chatbot com um banco de dados para armazenar as respostas e analisar os resultados.
Lembre-se de que este é um exemplo básico. Um chatbot completo requer mais funcionalidades e integrações, dependendo das necessidades específicas.


```python
def chatbot_welovepizza():

  # Mensagens iniciais
  print("Olá! Agradecemos sua visita à WeLovePizza. Queremos saber sua opinião para melhorarmos ainda mais!")
  print("Respondendo a esta pesquisa rápida, você concorre a 3 vinhos de sua escolha dentre uma seleção especial. 🍷")
  print("Para conferir as opções, acesse este link: [Link para a carta de vinhos]\n")

  # Perguntas e opções de resposta
  perguntas = {
      "experiencia": "Em uma escala de 0 a 10, sendo 0 “nada satisfeito” e 10 “extremamente satisfeito”, como você avalia sua experiência geral com a WeLovePizza hoje?",
      "pontos_fortes": "O que você mais gosta na WeLovePizza? (Pode escolher mais de uma opção)\n a) Sabor das pizzas\n b) Variedade do cardápio\n c) Qualidade dos ingredientes\n d) Preço\n e) Atendimento\n f) Tempo de entrega\n g) Ambiente da loja\n h) Outro: __________",
      "pontos_a_desenvolver": "O que poderíamos melhorar na WeLovePizza? (Pode escolher mais de uma opção)\n a) Sabor das pizzas\n b) Variedade do cardápio\n c) Qualidade dos ingredientes\n d) Preço\n e) Atendimento\n f) Tempo de entrega\n g) Ambiente da loja\n h) Outro: __________",
      "frequencia_compra": "Nos últimos 3 meses, com que frequência você comprou pizzas?\n a) Somente na WeLovePizza\n b) Mais na WeLovePizza do que na concorrência\n c) Mais na concorrência do que na WeLovePizza\n d) Somente na concorrência",
      "sentimento": "Qual palavra melhor descreve o que você sente pela WeLovePizza?\n a) Satisfação\n b) Confiança\n c) Indiferença\n d) Frustração\n e) Outro: __________",
      "sugestoes": "Tem alguma sugestão para melhorarmos ainda mais?"
  }

  # Loop para exibir as perguntas sequencialmente
  for key, pergunta in perguntas.items():
    if key == "experiencia":  # Validação específica para a questão 1
      while True:
        try:
          resposta = int(input(pergunta + "\nResposta (0-10): "))
          if 0 <= resposta <= 10:
            break
          else:
            print("Por favor, digite um número entre 0 e 10.")
        except ValueError:
          print("Entrada inválida. Digite um número.")
    else:  # Validação para as demais questões (com múltiplas escolhas)
      while True:
        resposta = input(pergunta + "\nResposta (separe as opções por ponto e vírgula para mais de uma resposta para a mesma pergunta): ").lower()
        opcoes_escolhidas = resposta.split(";")  # Separar as opções
        opcoes_validas = ["a", "b", "c", "d", "e", "f", "g", "h"]
        if all(opcao in opcoes_validas for opcao in opcoes_escolhidas):  # Verificar se todas as opções são válidas
           break
        else:
              print("Opção inválida. Por favor, escolha letras de 'a' a 'h', separadas por ponto e vírgula.")

    print()  # Linha em branco para separar as perguntas


  # ... (Mensagem final)
  print("Muito obrigado pela sua participação! Você já está concorrendo aos 3 vinhos. O resultado será divulgado em [data] em nossas redes sociais.")
```

```python
  # Execução da simulação
chatbot_welovepizza()
```
